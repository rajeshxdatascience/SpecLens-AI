```python
import pandas as pd
import numpy as np
```

```python
pd.set_option('display.max_columns',None)
pd.set_option('display.width',None)
pd.set_option('display.max_colwidth',None)
```

```python
df = pd.read_csv('/content/7_jan_phones_with_all_details.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df['brand'] = df['model'].str.lower().str.split(' ').str.get(0)
```

```python
df['brand'].value_counts()
```

```python
df.loc[df['brand'].str.contains('moto'),'brand'] = 'motorola'
```

```python
df['brand'].value_counts()
```

```python
df['brand'].count()
```

```python
df['brand'] = df['brand'].astype('category')
```

```python
df['beebome_score'].value_counts()
```

```python
df['rating'].value_counts()
```

```python
df['rating_count'] = df['rating_count'].str.lower().str.split('r').str.get(0)
```

```python
df['rating_count'] = df['rating_count'].str.replace('(','')
```

```python
df['rating_count'] = df['rating_count'].str.replace('k',"1000")
```

```python
df['rating_count'] = df['rating_count'].astype('float64')
```

```python
df.info()
```

```python
df['market_status'].value_counts()
```

```python
def market(text):

  t = text.lower()

  if t == 'in stock':
    return 1
  elif t == 'out of stock':
    return 0

df['is_available'] = df['market_status'].apply(market)
```

```python
df['is_available'].value_counts()
```

```python
df.drop(columns='market_status',inplace=True)
```

```python
df['launch_on']
```

```python
df['launch_on'] = pd.to_datetime(df['launch_on'])
```

```python
df['launch_on']
```

```python
df.info()
```

```python
df['old_price'].value_counts()
```

```python
df['old_price'] = df['old_price'].str.lower().str.split(' ').str.get(1)
```

```python
df.loc[df['old_price'] == 'price','old_price'] = np.nan
```

```python
df['old_price'].value_counts(dropna=False)
```

```python
df['old_price'] = df['old_price'].str.replace('₹','')
```

```python
df['old_price'] = df['old_price'].str.replace(',','')
```

```python
df['old_price'] = df['old_price'].astype('float')
```

```python
df.info()
```

```python
df['price'] = df['price'].str.replace('₹','')
```

```python
df['price'] = df['price'].str.replace(',','')
```

```python
df['price'] = df['price'].astype('int')
```

```python
df.info()
```

```python
df['launch_price'] = df['launch_price'].str.replace('₹','').str.replace(',','').astype(int)
```

```python
df.info()
```

```python
df['is_flipkart_available'] = df['flipkart_buy_link'].notna().astype('int')
```

```python
df['is_flipkart_available'].value_counts()
```

```python
df['is_amazon_available'] = df['amazon_buy_link'].notna().astype('int')
```

```python
df['is_amazon_available'].value_counts()
```

```python
df['is_official_available'] = df['official_store_link'].notna().astype('int')
```

```python
df['is_official_available'].value_counts()
```

```python
df.drop(columns=['flipkart_buy_link','amazon_buy_link','official_store_link'],inplace=True)
```
